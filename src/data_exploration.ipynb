```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#loading in data
village_LANDSAT = pd.read_csv('../Data/village_LANDSAT.csv')
```

```python
#seperating damaged and undamaged buildings
damaged = village_LANDSAT[village_LANDSAT['STATUS'] == 'DAMAGED']
undamaged = village_LANDSAT[village_LANDSAT['STATUS'] == 'NO DAMAGE']

#columns to use
cols = village_LANDSAT.columns[2:len(village_LANDSAT.columns)-1]

#plotting

for col in cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(damaged[col], color='red', label='Damaged', kde=True)
    sns.histplot(undamaged[col], color='blue', label='Undamaged', kde=True)
    plt.title(f'{col} damaged vs undamaged')
    plt.legend()
    plt.show()
```
